# LDA

```python
import gensim
from gensim import corpora
from gensim.models import CoherenceModel
from tqdm import tqdm
from nltk.tokenize import word_tokenize
```

```python
data['Tokenized'] = data['Deskripsi_Cleaned'].apply(lambda x: word_tokenize(x))
print(data[['Deskripsi_Cleaned', 'Tokenized']])
```

```python
id2word = corpora.Dictionary(data['Tokenized'])
corpus = [id2word.doc2bow(text) for text in data['Tokenized']]
```

```python
num_topics_range = range(5, 30, 1)
alpha_range = ['symmetric', 'asymmetric', 0.01, 0.31, 0.61, 0.91]
eta_range = ['symmetric', 0.01, 0.31, 0.61, 0.91]
```

```python
results = []

for num_topics in tqdm(num_topics_range):
    for alpha in alpha_range:
        for eta in eta_range:
            lda_model = gensim.models.LdaModel(
                corpus=corpus,
                id2word=id2word,
                num_topics=num_topics,
                random_state=100,
                update_every=1,
                chunksize=10,
                passes=10,
                alpha=alpha,
                eta=eta,
                per_word_topics=True
            )
            
            coherence_model_lda = CoherenceModel(model=lda_model, texts=data['Tokenized'], dictionary=id2word, coherence='c_v')
            coherence_score = coherence_model_lda.get_coherence()
            results.append({
                'num_topics': num_topics,
                'alpha': alpha,
                'eta': eta,
                'coherence_score': coherence_score
            })

results_df = pd.DataFrame(results)
print("Top 10 combinations based on Coherence Score:")
print(results_df.nlargest(10, 'coherence_score'))
```
